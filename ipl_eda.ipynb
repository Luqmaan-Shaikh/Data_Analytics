```python
# loading req libraries

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# load dataset

ipl = pd.read_csv("matches.csv")
```

```python
# first five records 

ipl.head()
```

```python
# looking at no of rows and columns in dataset

ipl.shape
```

```python
# freq of most MOTM Awards

ipl['player_of_match'].value_counts()
```

```python
# top 10 players with most MOTM Awards

ipl['player_of_match'].value_counts()[0:10]
```

```python
# top 5 players with most MOTM Awards

ipl['player_of_match'].value_counts()[0:5]
```

```python
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
# bar - plot for top 5 players with most MOTM Awards (Since data is categorical and numerical)

plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()), list(ipl['player_of_match'].value_counts()[0:5]), color= 'g')
plt.show()
```

```python
# freq of result col

ipl['result'].value_counts()
```

```python
# no toss win w.r.t each teams

ipl['toss_winner'].value_counts()
```

```python
# extracting records where a team won batting first

batting_first = ipl[ipl['win_by_runs'] != 0]
```

```python
batting_first.head()
```

```python
# making histogram (since data is numerical)

plt.figure(figsize=(7,7))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution of Runs")
plt.xlabel("Runs")
plt.show()
```

```python
# no of wins w.r.t each team after batting first

batting_first["winner"].value_counts()
```

```python
# bar plot for top 3 teams with most wins after batting first

plt.figure(figsize=(7,7))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()), list(batting_first['winner'].value_counts()[0:3]), color = ['blue', 'yellow', 'red'])
plt.show()
```

```python
# pie chart 

plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()), labels = list(batting_first['winner'].value_counts().keys()), autopct="%0.1f%%")
plt.show
```

```python
# batting second

batting_second = ipl[ipl['win_by_wickets'] != 0]
```

```python
batting_second.head()
```

```python
# histogram for freq of wins w.r.t no of wickets

plt.figure(figsize=(7,7))
plt.hist(batting_second['win_by_wickets'], bins = 30)
plt.xlabel("Wickets")
plt.show()
```

```python
# freq of no of wins w.r.t each team after batting second

batting_second['winner'].value_counts()
```

```python
# bar plot for top 3 teams with most wins after batting second

plt.figure(figsize=(7,7))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()), list(batting_second['winner'].value_counts()[0:3]), color = ['purple', 'blue', 'red'])
plt.show
```

```python
# pie chart for distribution of most wins after batting second

plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()), labels = list(batting_second['winner'].value_counts().keys()), autopct='%0.1f%%')
plt.show
```
